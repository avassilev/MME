Разгледайте задачата за максимизация на печалбата на монопол, който оперира на пазар, описван с обратната функция на търсене

$$ p(q) = A q^{-\alpha} .$$

- Реализирайте функцията на обратно търсене на Julia. Използвайте параметри $A=2.0$ и $\alpha = 0.3$. Визуализирайте функцията на обратно търсене.
- Използвайте JuMP и намерете числено оптималното количество, при което се максимизира печалбата на монополиста, ако се използва функцията на обратно търсене от предходната подточка и функцията на общи разходи $C(q) = f + v q^2$, където $f=0.1$ и $v=1.0$.
- Илюстрирайте визуално получения резултат, като изобразите на една графика функциите на общите приходи $R(q)$ и на общите разходи $C(q)$, както и вертикална права за изчислената стойност за оптималното количество.
